```python
# diagram_batang.py
import matplotlib.pyplot as plt

# Data
siswa = ["Andi", "Budi", "Citra", "Dewi", "Eka"]
nilai = [80, 65, 90, 75, 85]

# Buat diagram batang
plt.figure(figsize=(8, 5))
plt.bar(siswa, nilai, color="skyblue", edgecolor="black")

# Tambahkan detail
plt.title("Diagram Batang Nilai Ulangan Matematika")
plt.xlabel("Nama Siswa")
plt.ylabel("Nilai")
plt.ylim(0, 100)  # batas sumbu Y
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Tampilkan
plt.show()

```
