```python
import numpy as np
import matplotlib.pyplot as plt

def plot_trigonometri_umum():
    print("=== GRAFIK FUNGSI TRIGONOMETRI UMUM ===")
    print("Format fungsi: a sin(bx) | a cos(bx) | a tan(bx)")
    fungsi = input("Pilih jenis fungsi (sin/cos/tan): ").strip().lower()

    # Input koefisien a dan b
    try:
        a = float(input("Masukkan nilai a (amplitudo): "))
        b = float(input("Masukkan nilai b (frekuensi): "))
    except ValueError:
        print("Nilai a dan b harus berupa angka.")
        return

    # Input batas sudut (derajat)
    try:
        start_deg = float(input("Masukkan batas awal sudut (derajat): "))
        end_deg   = float(input("Masukkan batas akhir sudut (derajat): "))
    except ValueError:
        print("Batas sudut harus berupa angka.")
        return

    if end_deg <= start_deg:
        print("Batas akhir harus lebih besar dari batas awal.")
        return

    # Konversi ke radian
    start_rad = np.radians(start_deg)
    end_rad   = np.radians(end_deg)

    # Sumbu x
    x = np.linspace(start_rad, end_rad, 1000)

    # Hitung y sesuai fungsi
    if fungsi == "sin":
        y = a * np.sin(b * x)
        judul = f"y = {a}sin({b}x)"
    elif fungsi == "cos":
        y = a * np.cos(b * x)
        judul = f"y = {a}cos({b}x)"
    elif fungsi == "tan":
        y = a * np.tan(b * x)
        # Hindari garis vertikal akibat tak hingga
        y[np.abs(y) > 10] = np.nan
        judul = f"y = {a}tan({b}x)"
    else:
        print("Jenis fungsi tidak dikenali. Gunakan sin, cos, atau tan.")
        return

    # Plot
    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label=judul, color='blue')

    # Garis sumbu
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)

    # Label & grid
    plt.title(f"Grafik {judul}  (dari {start_deg} sampai {end_deg})")
    plt.xlabel("x (radian)")
    plt.ylabel("y")
    plt.grid(True)
    plt.legend()

    # Tampilkan ticks sumbu x dalam derajat
    xticks_deg = np.linspace(start_deg, end_deg, 9)
    plt.xticks(
        np.radians(xticks_deg),
        [f"{int(d)}" for d in xticks_deg]
    )

    plt.show()

# Jalankan program
if __name__ == "__main__":
    plot_trigonometri_umum()

```
