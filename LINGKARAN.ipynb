```python
# lingkaran.py
import numpy as np
import matplotlib.pyplot as plt

# Jari-jari lingkaran
r = 3

# Buat sudut dari 0 sampai 2Ï€
theta = np.linspace(0, 2*np.pi, 500)

# Hitung koordinat x dan y
x = r * np.cos(theta)
y = r * np.sin(theta)

# Gambar lingkaran
plt.figure(figsize=(6,6))
plt.plot(x, y, color="blue", linewidth=2)
plt.title(f"Lingkaran dengan jari-jari {r}")
plt.xlabel("x")
plt.ylabel("y")

# Atur agar sumbu sama panjang (supaya lingkaran tidak jadi elips)
plt.axis("equal")
plt.grid(True)

# Tampilkan
plt.show()
```
